```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
# Step 1: Generate a 2D dataset
np.random.seed(42)
X = np.random.multivariate_normal(mean=[2, 5], cov=[[3, 2], [2, 2]], size=100)

# Step 2: Plot the original data
plt.scatter(X[:, 0], X[:, 1], alpha=0.7)
plt.title("Original 2D Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.axis("equal")
plt.grid(True)
plt.show()
```

```python
# Step 3: Center the data
X_centered = X - np.mean(X, axis=0)

# Step 4: Compute the covariance matrix
cov_matrix = np.cov(X_centered.T)

# Step 5: Compute eigenvalues and eigenvectors
eig_vals, eig_vecs = np.linalg.eig(cov_matrix)
```

```python
# Step 6: Project data onto the first principal component
pc1 = eig_vecs[:, 0]  # First principal component
X_pca_1d = X_centered @ pc1  # Projected data in 1D

# Step 7: Plot the 1D projection
plt.scatter(X_pca_1d, np.zeros_like(X_pca_1d), alpha=0.7)
plt.title("1D Projection onto First Principal Component")
plt.xlabel("Projected Value")
plt.grid(True)
plt.show()
```

```python

# Step 8: Plot variance explained
explained_variance = eig_vals / np.sum(eig_vals)
plt.bar([1, 2], explained_variance, tick_label=["PC1", "PC2"])
plt.title("Variance Explained by Each Principal Component")
plt.ylabel("Explained Variance Ratio")
plt.grid(True)
plt.show()
```
